```
import cv2
from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('./dog1.jpg')
print(img.shape)
c, r = img.shape[:2]
img = cv2.resize(img, (int(0.3 * c), int(0.3 * r)), cv2.INTER_AREA)
cv2_imshow(img)
```

```
img_temp = cv2.imread('./dog1_temp.JPG')
cv2_imshow(img_temp)
```

```
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_temp = cv2.cvtColor(img_temp, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
cv2_imshow(gray_temp)
```

```
w, h = gray_temp.shape
```

```
output = cv2.matchTemplate(gray, gray_temp, cv2.TM_CCOEFF_NORMED)
cv2.minMaxLoc(output)
```

```
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(output)
min_val
```

```
top = max_loc # 직사각형 왼쪽 위 코너 좌표
top
```

```
bottom = (top[0] + h, top[1] + w)
cv2.rectangle(img, top, bottom, 255, 2)
cv2_imshow(img)
```

```
img = cv2.imread('./dog1.jpg')
print(img.shape)
c, r = img.shape[:2]
img = cv2.resize(img, (int(0.3 * c), int(0.3 * r)), cv2.INTER_AREA)
cv2_imshow(img)

img_temp = cv2.imread('./dog1_temp.JPG')
# cv2_imshow(img_temp)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_temp = cv2.cvtColor(img_temp, cv2.COLOR_BGR2GRAY)
# cv2_imshow(gray)
# cv2_imshow(gray_temp)

w, h = gray_temp.shape
output = cv2.matchTemplate(gray, gray_temp, cv2.TM_CCOEFF_NORMED)
cv2.minMaxLoc(output)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(output)
top = max_loc # 직사각형 왼쪽 위 코너 좌표

bottom = (top[0] + h, top[1] + w)
cv2.rectangle(img, top, bottom, 255, 2)
cv2_imshow(img)
```
